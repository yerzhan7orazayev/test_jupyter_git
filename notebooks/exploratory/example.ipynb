```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(500)

plt.figure()
plt.hist(data)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
